```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3= pd.read_csv(url3)
```

```python
df1
```

```python
df2
```

```python
df3

```

```python
df1.shape
df2.shape
df3.shape
```

```python
print("Columns in DataFrame 1:", df1.columns)
print("Columns in DataFrame 2:", df2.columns)
print("Columns in DataFrame 3:", df3.columns)
```

```python
df3 = df3.rename(columns={'State': 'ST', 'Gender': 'GENDER'})
```

```python
example_column_order = ['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']
```

```python
df2 = df2[example_column_order]
df3 = df3[example_column_order]
```

```python
print("Columns in DataFrame 1:", df1.columns)
print("Columns in DataFrame 2:", df2.columns)
print("Columns in DataFrame 3:", df3.columns) #I brought the column in the same order
```

```python
concat_df = pd.concat([df1, df2 , df3])
```

```python
concat_df.shape
concat_df.columns = [column.upper() for column in concat_df.columns]
concat_df.columns = [column.replace(' ','_') for column in concat_df.columns]
concat_df.columns
```

```python
#to clean "GENDER" coulumn :
concat_df["GENDER"].unique()
```

```python
concat_df["GENDER"] = concat_df["GENDER"].replace({"Femal":"F", "Male" : "M" , "female" :"F"})
```

```python
concat_df["GENDER"].unique()  #Now the gender colomn has just M; F and Nan
```

```python
#to clean "ST" coulumn :
concat_df["ST"].unique()
```

```python
concat_df["ST"] =concat_df["ST"].replace({"AZ":"Arizona", "Cali":"California", "WA": "Washington",})
```

```python
concat_df["ST"].unique()
```

``>to clean "Education" column``

```python

concat_df["EDUCATION"].unique()
```

```python
concat_df["EDUCATION"] =concat_df["EDUCATION"].replace({"Bachelors":"Bachelor"})
```

``>to clean "Vehicle_Class" column``

```python
concat_df["VEHICLE_CLASS"] = concat_df["VEHICLE_CLASS"].replace({"Sports Car":"Luxury","Luxury SUV":"Luxury","Luxury Car":"Luxury"})
```

```python
concat_df["VEHICLE_CLASS"].unique()
```

```python
concat_df['CUSTOMER_LIFETIME_VALUE'] = concat_df['CUSTOMER_LIFETIME_VALUE'].str.replace('%', '').astype(float)
```

for formating data types:

```python
concat_df['MONTHLY_PREMIUM_AUTO'] = pd.to_numeric(concat_df['MONTHLY_PREMIUM_AUTO'], errors='coerce')
concat_df['INCOME'] = pd.to_numeric(concat_df['INCOME'], errors='coerce')

print(concat_df.dtypes)
```

```python
concat_df['NUMBER_OF_OPEN_COMPLAINTS'].unique() 
```

```python


# Replace specific string values with corresponding integers
concat_df["NUMBER_OF_OPEN_COMPLAINTS"] = concat_df["NUMBER_OF_OPEN_COMPLAINTS"].replace({
    "1/0/00": 0,
    "1/1/00": 1,
    "1/3/00": 3,
    "1/5/00": 5,
    "1/4/00": 4,
    "1/2/00": 2
})

# Convert the column to integers
concat_df["NUMBER_OF_OPEN_COMPLAINTS"] = concat_df["NUMBER_OF_OPEN_COMPLAINTS"].astype(int)

```

```python
concat_df.dropna(axis=0, how='all', inplace=True)
print(concat_df)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
pd.options.display.max_columns = None

url_challenge2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url_challenge2)
df
```

```python
df_grouped = df.groupby(["sales_channel"])[["monthly_premium_auto"]].sum()
df_grouped.T
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df_pivot = df.pivot_table(values= ["monthly_premium_auto"],columns = ["sales_channel"], aggfunc = "sum")

df_pivot 

```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df_grouped2 = df.groupby(["gender", "education"])[["customer_lifetime_value"]].mean()
df_grouped2
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
